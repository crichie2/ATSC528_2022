```python
### Import Required Modules (shouldn't need to change) ###
import numpy as np                 #numpy for math
import matplotlib.pyplot as plt    #matplotlib for plotting
import matplotlib.image as mpimg   #for funsies
import cartopy.crs as ccrs         #cartopy for plotting on map
import cartopy.feature as cfeature #cartopy basic shapefiles
```

```python
# Base conditions for equation building # 

t_step = 4/(24*5)
a = 0.25
b = 4.0
F = 8.0
G = 1.0
```

```python
# Building a function to run timesteps # 

def timestep(x,y,z):
    x_step = x + t_step*( -y**2 - z**2 - a*x + a*F)
    y_step = y + t_step*( x*y - b*x*z - y + G)
    z_step = z + t_step*( b*x*y + x*z - z)
    return ([x_step, y_step, z_step])
```

```python
# List building and loop building to run the equations over time # 

# Lists for initial conditions #
initx = np.empty([720])
inity = np.empty([720])
initz = np.empty([720])

# Setting initial conditions for run #
initx[0] = 2.5
inity[0] = 1 
initz[0] = 0

# This for loop will run the dimensions for the timesteps out to 720 hours # 

for i in range(1, 720):
        step = timestep(initx[i-1], inity[i-1], initz[i-1])
        initx[i] = step[0]
        inity[i] = step[1]
        initz[i] = step[2]
        
        #print(step)
```

```python
#Same as above, but different inital conditions #

# Lists for Second conditions #
secondx = np.empty([720])
secondy = np.empty([720])
secondz = np.empty([720])

# Setting conditions for second run #
secondx[0] = 2.49
secondy[0] = 1
secondz[0] = 0


for i in range(1, 720):
        step2 = timestep(secondx[i-1], secondy[i-1], secondz[i-1])
        secondx[i] = step2[0]
        secondy[i] = step2[1]
        secondz[i] = step2[2]
        
        #print(step2)
```

```python
# Plot that graph up X edition # 
# Also it is Army-Navy week, so Navy Blue and Gold plots #

x_axis = np.arange(0,720,1)

fig = plt.figure(figsize=(15,9))
plt.plot(x_axis, initx, color = 'navy', label = 'Run 1')
plt.plot(x_axis, secondx, color = 'gold', label = 'Run 2')
plt.title('X results: Go NAVY! Beat Army!')
plt.savefig('X Results.png')
plt.legend(loc = 'upper right')
```

```python
# Plot that graph up Y edition # 
x_axis = np.arange(0,720,1)
fig = plt.figure(figsize=(15,9))
                 
            
plt.plot(x_axis, inity, color = 'navy', label = 'Run 1')
plt.plot(x_axis, secondy, color = 'gold', label = 'Run 2')
plt.title('Y results')
plt.savefig('Y Results.png')
plt.legend(loc = 'upper right')
```

```python
# Plot that graph up Z edition # 
x_axis = np.arange(0,720,1)
fig = plt.figure(figsize=(15,9))

plt.plot(x_axis, initz, color = 'navy', label = 'Run 1')
plt.plot(x_axis, secondz, color = 'gold', label = 'Run 2')
plt.title('Z results')
plt.savefig('Z Results.png')
plt.legend(loc = 'upper right')
```
